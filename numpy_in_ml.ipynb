```python
matrix1 = np.array([
    [1, 2, 3],
    [4, 5, 6],
])

matrix2 = np.array([
    [5, 6, 7],
    [5, 6, 7],
])

vector = [1, 2, 3]

print(matrix1 + matrix2)
print(matrix1 - matrix2)
print(matrix1 * matrix2)
print((matrix1 / matrix2).round(2))
print(matrix1 + 2)
print(matrix1 - 2)
print(matrix1 * 2)
print(matrix1 / 2)
print(matrix1 * vector)
print(np.dot(matrix1, vector))
print(matrix1.T)
```
